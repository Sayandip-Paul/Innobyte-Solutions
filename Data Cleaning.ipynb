## Importing the necessary libraries

```python
import numpy as np
import pandas as pd
```

## Loading the Data

```python
df = pd.read_csv('./Amazon_Sale_Report.csv')
```

## Displaying the Data

```python
df.head()
```

## Missing Values

```python
df.isnull().sum()
```

```python
df.info()
```

## Dropping the Null Columns

```python
df.drop(columns = ['New','PendingS'],inplace = True)
```

## Filling the missing currency values

```python
df['currency'] = df['currency'].fillna('INR')
```

```python
df.isnull().sum()
```

## Filling the missing Amount values with zero

```python
df['Amount'] = df['Amount'].fillna(0.0)
```

```python
df.isnull().sum()
```

## Dropping the missing values in ship-city, ship-state, ship-postal-code, ship-country

```python
df.dropna(subset = ['ship-city','ship-state','ship-postal-code','ship-country'],inplace = True)
```

```python
df.isnull().sum()
```

```python
df['fulfilled-by'] = df['fulfilled-by'].fillna('Easy Ship')
```

```python
df.isnull().sum()
```

```python
df.info()
```

```python
df.head()
```

```python
df['Date'].value_counts() ## MM-DD-YY
```

## Converting the date column into pandas datetime in dd-mm-yyyy format

```python
df['Date'] = pd.to_datetime(df['Date']).dt.strftime('%d-%m-%Y')
```

```python
df.head()
```

## Exporting the data as a csv file

```python
df.to_csv('amazon_sales_clean.csv',index = False)
```
